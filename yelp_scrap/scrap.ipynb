```python
import csv
import os
import pandas as pd
import json

# Specify the directory where your CSV files are located
json_directory = 'city_business_json'
csv_directory = 'city_business_csv'

# Loop through all files in the directory
for filename in os.listdir(json_directory):
    if filename.endswith('.json'):          
        # Read the original JSON file
        # city_business_json//Michigan_yelp_scrap_dataset_business.json
        input_file_path = os.path.join(json_directory, filename)
        print(input_file_path)
        with open(input_file_path, 'r') as json_file:
            data = json_file.read()

        # Remove line breaks and extra spaces
        formatted_data = ''.join(data.split())
        formatted_data = formatted_data.replace('}{', '}\n{')

        with open('output.json', 'w') as json_file:
            json_file.write(formatted_data)


        # Initialize an empty list to store the JSON objects
        yelp_business = []

        # Open the JSON file for reading
        with open("output.json", 'r', encoding='utf-8') as yelp_business_file:
            for line in yelp_business_file:
                # Load each JSON object and append it to the list
                entry = json.loads(line)
                yelp_business.append(entry)

        # Convert the list of dictionaries to a pandas DataFrame for more advanced data analysis
        yelp_business_df = pd.DataFrame(yelp_business)
        yelp_business_df = yelp_business_df.dropna(subset=['id'], how='any')
        yelp_business_df = yelp_business_df.drop_duplicates(subset=['id'], keep='first')
        
        output_file_path = os.path.join(csv_directory, filename)[:-4]+"csv" 
        print(output_file_path)
        yelp_business_df.to_csv(output_file_path, index = False)
        
```
